```python
def cleaner(df,columns_):
    # Getting the nation of the player 
    df['Nation'] = df['Nation'].apply(lambda x  : str(x).split()[-1])
    
    # Renaming the columns
    df.columns = columns_

    # deleting the Matches Columns
    del df['Matches']
    return df
```
