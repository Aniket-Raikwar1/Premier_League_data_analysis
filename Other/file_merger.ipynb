```python
import numpy as np
import pandas as pd
```

```python
standard = 'Datasets/Players_data/2023-2024/Player_stats_standard-2023-2024'
misc = 'Datasets/Players_data/2023-2024/Player_stats_misc-2023-2024'
passing_t = 'Datasets/Players_data/2023-2024/Player_stats_passing_types-2023-2024'
passing = 'Datasets/Players_data/2023-2024/Player_stats_passing-2023-2024'
playingtime = 'Datasets/Players_data/2023-2024/Player_stats_playing_time-2023-2024'
posses = 'Datasets/Players_data/2023-2024/Player_stats_possession-2023-2024'
shoot = 'Datasets/Players_data/2023-2024/Player_stats_shooting-2023-2024'
defe = 'Datasets/Players_data/2023-2024/Player_stats_defense-2023-2024'
gca = 'Datasets/Players_data/2023-2024/Player_stats_gca-2023-2024'
keep_ad= 'Datasets/Players_data/2023-2024/Player_stats_keeper_adv-2023-2024'
keep = 'Datasets/Players_data/2023-2024/Player_stats_keeper-2023-2024'

df_stand = pd.read_csv(standard)
df_misc = pd.read_csv(misc)
df_passing_t = pd.read_csv(passing_t)
df_passing = pd.read_csv(passing)
df_playingtime = pd.read_csv(playingtime)
df_posses = pd.read_csv(posses)
df_shoot = pd.read_csv(shoot)
df_defe = pd.read_csv(defe)
df_gca = pd.read_csv(gca)
df_keep_ad = pd.read_csv(keep_ad)
df_keep = pd.read_csv(keep)
```

```python
'''
keeper adv - 203, keeper - 203,
misc - 2852,  defense - 2852, gca - 2852, passing - 2852, passing type - 2852, possesion - 2852, shooting - 2852, standard - 2852,
playing time - 3512,
'''
```

```python
def player_id(df):
    df['Player_id'] = df['Player'].apply(lambda x : "_".join(str(x).split(" "))) + "_" +df['Squad'].apply(lambda x : "_".join(str(x).split(" ")))
    return df
```

```python
df_playingtime = player_id(df_playingtime)
df_defe = player_id(df_defe) # Merged
df_gca = player_id(df_gca) # Merged
df_keep = player_id(df_keep)
df_keep_ad = player_id(df_keep_ad)
df_misc = player_id(df_misc)# Merged
df_passing = player_id(df_passing) # Merged
df_passing_t = player_id(df_passing_t) # Merged
df_posses = player_id(df_posses) # Merged
df_shoot = player_id(df_shoot) # Merged
df_stand = player_id(df_stand) # Merged
```

```python
df_merged = df_stand.merge(
    df_shoot[['Goals', 'Shots', 'Shots on Target',
       'Shots on Target %', 'Shots per 90', 'Shoots on Target per 90',
       'Goal per shot', 'Goals per shot on Target', 'Average shot dist.',
       'Free Kicks', 'Penalty made', 'Penalty attempted', 'Expected Goals xG',
       'Non-Penalty Expected Goals npxG',
       'Non-Penalty Expected goals per shot npxG/Sh',
       'Goals minus Expected Goals G-xG',
       'Non-penalty goals minus non-penalty Expected goals np:G-xG',
       'Player_id']],
    how='left',on='Player_id'
)
```
